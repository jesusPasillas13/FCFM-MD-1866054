Ejercicio Listas

```python
import random
r=range(1,250)
lista1=random.sample(r,40)
print(lista1)

def par(lista):
    par=list
    for x in lista:
        if x%2 == 0:
            par.append(x)
    return par
def impar(lista):
    impar=list
    for x in lista:
        if x%2 != 0:
            impar.append(x)
    return impar
listap=par(lista1)
print(listap)
listai=impar(lista1)
print(listai)

print("")
lp=len(listap)
li=len(listai)
print("Cantidad de pares: ",lp, "Cantidad de impares: ",li)

print("")
listai.sort()
listap.sort(reverse=True)
print("Pares: ",listap)
print("Impares: ",listai)

print("")
print("Longitud de lista: ", len(lista1))
print("Suma de los valores: ", sum(lista1))
print(" Numero menor de la lista: ", min(lista1))
print("Numero mayor de la lista: ", max(lista1))
```
